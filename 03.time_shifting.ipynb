# Time shifting

https://www.programiz.com/python-programming/pandas/methods/shift

```python
import pandas as pd

df = pd.read_csv("starbucks.csv",index_col = "Date", parse_dates=True)
df.head()
```

```python
df.shift(1)  #shift forward(end up losng one entry at the last)
```

```python
df.shift(5)
```

```python
df.shift(-1)  #shift backward(start up losing one entry at the first)
```

```python
df.shift(periods = 1,freq='M') #shift by month
```
